```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset from a CSV file into a DataFrame
file_path = "../Dataset/nba_salaries.csv"
df = pd.read_csv(file_path)
```

```python
# Let's begin by exploring the dataset
# Display the first few rows of the DataFrame
df.head()
```

```python
# Check the dimensions of the DataFrame (rows, columns)
df.shape
```

```python
# Check the data types of each column
df.dtypes
```

```python
# Check the data types of each column
df.dtypes
```

```python
# Perform summary statistics on numerical columns
df.describe()
```

```python
# checking unique values
df.nunique()
```

```python
# Check the unique teams in the dataset
unique_teams = df['Team'].unique()
unique_teams
```

```python
# checking null values
df.isnull().sum()
```

```python
# percentage of null values
df.isnull().sum()/ len(df) *100 
```

```python
# checking columns
df.columns
```

```python
# Find the highest-paid player and their salary
highest_paid_player = df[df['Salary'] == df['Salary'].max()]
```

```python
# checking the player
highest_paid_player
```

```python
# Calculate the highest-paid player for each team
highest_paid_players = df.groupby('Team').apply(lambda x: x.loc[x['Salary'].idxmax()])
highest_paid_players
```

```python
# Group the data by team and calculate the average salary for each team
team_average_salary = df.groupby('Team')['Salary'].mean()
```

```python
# checking average salary of team
team_average_salary
```

```python
# Group the data by player position and calculate the average salary for each position
position_average_salary = df.groupby('Position')['Salary'].mean()
```

```python
# Calculate the number of players in each position
position_counts = df['Position'].value_counts()
position_counts
```

```python
# Create a heatmap to visualize correlations between numeric variables
correlation_matrix = df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of NBA Player Statistics')
plt.show()
```

```python
# Create a countplot to visualize the distribution of player positions
plt.figure(figsize=(10, 6))
sns.countplot(x='Position', data=df, palette='Set2')
plt.xlabel('Position')
plt.ylabel('Count')
plt.title('Distribution of NBA Player Positions')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a barplot to show the average salary by team
plt.figure(figsize=(12, 6))
sns.barplot(x='Team', y='Salary', data=df, estimator=np.mean, ci=None, palette='viridis')
plt.xlabel('Team')
plt.ylabel('Average Salary')
plt.title('Average Salary by NBA Team')
plt.show()
```

```python
# Create a histplot to visualize the distribution of player ages
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of NBA Player Ages')
plt.show()
```

```python
# Create a pie chart to show the distribution of players by team
plt.figure(figsize=(8, 8))
team_counts = df['Team'].value_counts()
plt.pie(team_counts, labels=team_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of NBA Players by Team')
plt.axis('equal')
plt.show()
```

```python
# Create a histogram of player salaries
plt.figure(figsize=(10, 6))
sns.histplot(df['Salary'], bins=20, kde=True, color='purple')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Distribution of NBA Player Salaries')
plt.show()
```

```python
# Create a boxplot to compare salaries by player position
plt.figure(figsize=(12, 6))
sns.boxplot(x='Position', y='Salary', data=df, palette='Set3')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.title('Salary Distribution by NBA Player Position')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a scatterplot to visualize the relationship between age and salary
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', data=df, hue='Position', palette='hot')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Scatter Plot of NBA Player Age vs. Salary')
plt.show()
```

```python
df.columns
```

```python
# Create a pairplot to explore relationships between numeric variables
sns.pairplot(df[['Age', 'Salary', 'Position']], hue='Position', palette='viridis')
plt.suptitle('Pairplot of NBA Player Statistics', y=1.02)
plt.show()
```

```python
# Create a swarmplot to visualize salary distribution by team
plt.figure(figsize=(12, 8))
sns.swarmplot(x='Team', y='Salary', data=df, palette='husl')
plt.xlabel('Team')
plt.ylabel('Salary')
plt.title('Salary Distribution by NBA Team')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a catplot to show the distribution of positions within each team
sns.catplot(x='Team', hue='Position', data=df, kind='count', palette='muted', height=8, aspect=2)
plt.xlabel('Team')
plt.ylabel('Count')
plt.title('Distribution of NBA Player Positions by Team')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create an lmplot to visualize the linear relationship between age and salary
sns.lmplot(x='Age', y='Salary', data=df, height=6, aspect=2)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Linear Relationship Between NBA Player Age and Salary')
plt.show()
```
